```python
i_list = range(len(df)-1)
j_list = range(len(df)-1)
k_list = range(N-1)

print(i_list)
print(j_list)
print(k_list)
```

### Objective function

$Min \; Z= \; \displaystyle\sum_{j=1}^{n}C_{0j}\sum_{k=1}^{N} Ky^{k}_{0j}+\sum_{i=1}^{n}\sum_{j=1, j\neq1}^{n}C_{ij}\sum_{k=1}^{N-1} Ky^{k}_{ij}$

```python
mdl.minimize(mdl.sum( df[0][j] * mdl.sum(k * x[0,j,k] for k in k_list) for j in j_list if j!=0 ) + mdl.sum(df[i][j] *mdl.sum(k* x[i,j,k] for k in k_list)for j in j_list for i in i_list if j!=i ))
```

$\displaystyle\sum_{k=1}^{N} y^{k}_{0j}+\sum_{k=1}^{N-1}\sum_{i=1, i\neq j}^{n} y^{k}_{ij} = 1$     $\; \; (\forall \;j \; \in \; I)$

$\displaystyle\sum_{j=1}^{N} y^{1}_{0j}+\sum_{i=1}^{n}\sum_{j=1, j\neq i}^{n} y^{1}_{ij} = m$     $\; \; (\forall \;i \; \in \; I)$

$\displaystyle\sum_{k=1}^{N} \sum_{j=1}^{n} y^{k}_{0j}=m $

$\displaystyle y^{k+1}_{0j} +  \sum_{i=1, i\neq j}^{n} y^{k+1}_{ij} = \sum_{i=1, i\neq j}^{n} y^{k}_{ji} $     $\; \; (\forall \;i \; \in \; I ; k=1,2,...,N-2)$

$\displaystyle y^{N}_{0j} = \sum_{i=1, i\neq j}^{n} y^{N-1}_{ji} $     $\; \; (\forall \;j \; \in \; I)$

```python
# Restricciones:

#1
for j in j_list:
    mdl.add_constraint((mdl.sum(x[0,j,k] for k in k_list if j!=0)+mdl.sum(x[i, j, k] for k in k_list for i in i_list if i!=j)) ==1)
                      
#2
for i in i_list:
    mdl.add_constraint((mdl.sum(x[0,j,1] for j in j_list if j!=0)+mdl.sum(x[i,j,1] for i in i_list for j in j_list if j!=i)) == m)
                        
#3
    mdl.add_constraint((mdl.sum(x[0,j,k] for k in k_list for j in k_list if j!=0)) == m)
#4
for k in range(1,N-2):
    for i in i_list:
        mdl.add_constraint((x[0,j,k+1] + mdl.sum(x[i,j,k+1] for j in j_list if j!=i)) == mdl.sum(x[j,i,k] for i in i_list if i!=j) )

#5    
for j in range(1,len(df)-1):
    mdl.add_constraint((x[0,j,N-2]) == mdl.sum(x[j,i,N-1] for i in i_list if i!=j))
```

```python
print(mdl.export_to_string())
```

```python
#mdl.parameters.timelimit=120
#mdl.parameters.mip.strategy.branch=1
#mdl.parameters.mip.tolerances.mipgap=0.15

solucion = mdl.solve(log_output=True)
```
